### Importing all libraries required in this notebook

```
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Reading dataset

```
data = pd.read_csv("/content/SampleSuperstore.csv")
data.head()
```

Describing Data

```
data.info()
```

```
data.describe()
```

```
data.corr()
```

Data Visualization

```
data.hist(figsize=(15,10))
```

```
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```
sns.distplot(data['Sales'])
plt.xlim(-200,5000)
```

```
sns.distplot(data['Quantity'])
```

```
sns.countplot(x=data['Category'])
```

```
plt.bar('Category','Sales',data=data)
plt.title("Category vs Sales")
plt.xlabel('Category')
plt.ylabel('Sales')
```

```
sns.barplot(x="Region",y="Profit",data=data)
```

```
sns.countplot(x=data['Ship Mode'])
```

```
sns.countplot(x='Category',hue='Region',data=data)
plt.title("Category wise Region")
```

```
sns.barplot(x="Category",y="Profit",data=data)
```

```
plt.figure(figsize=(25,8))
sns.countplot(x=data['Sub-Category'])
```

```
plt.figure(figsize=(25,8))
sns.countplot(x='Sub-Category',hue='Region',data=data)
plt.title("Sub-Category wise Region")
```

```
plt.figure(figsize=(25,8))
sns.barplot(x="Sub-Category",y="Profit",data=data)
```

```
plt.figure(figsize=(25,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=60)
```

```
data.groupby(['Segment']).sum()[['Sales','Profit']].plot.bar()
plt.title("Segment wise Sales vs Profit")
```

```
data.groupby(['Ship Mode']).sum()[['Sales','Profit']].plot.bar()
plt.title("Ship Mode wise Sales vs Profit")
```

```
data.groupby(['Category']).sum()[['Sales','Profit']].plot.bar()
plt.title("Category wise Sales vs Profit")
```

```
data.groupby(['Sub-Category']).sum()[['Sales','Profit']].plot.bar()
plt.title("Sub-Category wise Sales vs Profit")
```

```
sns.barplot(x='Region',y='Sales',hue='Segment',data=data)
```
